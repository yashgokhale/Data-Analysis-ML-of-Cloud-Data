```python
def st_clustering(t1,h1,t2,h2):
    t_clust=[]
    h_clust=[]
    output_resolution=(len(t1),len(h1))
    data_resolution=(len(t2),len(h2))
    print('--------Start of clustering cycle--------')
    print(f'Changing time resolution from {len(t2)}->{len(t1)}')
    print(f'Changing height resolution from {len(h2)}->{len(h1)}')
    if len(t1)>len(t2):
        print('You are increasing time resolution')
        for i in t1:
            x=np.argmin(np.abs(i-t2))
            t_clust.append(x)
    else:
        print('You are decreasing time resolution')
        for i in range(len(t1)):
            if i!=len(t1)-1:
                ltemp=[m for m,x in enumerate(t2) if t1[i]<= x <=t1[i+1]]
                t_clust.append(ltemp)
            else:
                ltemp=[m for m,x in enumerate(t2) if t1[i]<= x]
                t_clust.append(ltemp) 
    if len(h1)>len(h2):
        print('You are increasing height resolution')
        for i in h1:
            x=np.argmin(np.abs(i-h2))
            h_clust.append(x)
    else:
        print('You are decreasing height resolution')     
        for i in range(len(h1)):
            if i!=len(h1)-1:
                ltemp=[m for m,x in enumerate(h2) if h1[i]<= x <=h1[i+1]]
                h_clust.append(ltemp)
            else:
                ltemp=[m for m,x in enumerate(h2) if h1[i]<= x]
                h_clust.append(ltemp) 
    print('--------End of clustering cycle--------')
    return t_clust,h_clust
```

```python
def val_collection(targs,hargs,val):
    start=time.time()
    name=val.long_name
    bad=val.missing_value
    val=np.ma.filled(val[:])
    u=[]
    if type(targs[0])==list and type(hargs[0])!=list:
        for t in targs:
            for h in hargs:
                u.append([val[i,h] for i in t])
    elif type(targs[0])==list and type(hargs[0])==list:
        for t in targs:
            for h in hargs:
                u.append([val[i,j] for i in t for j in h])
    elif type(targs[0])!=list and type(hargs[0])==list:
        for t in targs:
            for h in hargs:
                u.append([val[t,j] for j in h])
    elif type(targs[0])!=list and type(hargs[0])!=list:
        for t in targs:
            for h in hargs:
                u.append([val[t,h]])  
    end=time.time()
    mm=(end-start)
    print(f'Clustering of {name} took {mm:1.2f} s')
    return u,bad,name
```

```python
def val_filtering(targs,hargs,val):
    val,bad,name=val_collection(targs,hargs,val)
    start=time.time()
    av=[]
    st=[]
    for m in val:
        x=np.array(m)
        x=x[x!=bad]
        x=x[x!=0]
        av.append(np.average(x))
        st.append(np.std(x))
    av=np.reshape(av,(len(targs),len(hargs)))
    st=np.reshape(st,(len(targs),len(hargs)))
    end=time.time()
    mm=end-start
    print(f'Filtering of {name} took {mm:1.2f} s')
    return av,st
```

```python
def data_gen_par(targs,hargs,val,e_ind):
    start=time.time()
    av,st=val_filtering(targs,hargs,val)
    a1=[]
    s1=[]
    for i in range(len(av)):
        a1.append(av[i,int(e_ind[i])])
        s1.append(st[i,int(e_ind[i])])
    end=time.time()
    lll=end-start
    print(f'Total time: {lll}s')
    return a1,s1
```

```python
def extinction(ext,t1,h1,threshold=50,upper=100):
    base_h=np.zeros(len(t1))
    extin_h=np.zeros(len(t1))
    upper_extin=np.zeros(len(t1))
    lower_extin=np.zeros(len(t1))
    ht=np.zeros(len(t1))
    for i in range(len(ext)):
        em=ext[i]
        for k,j in enumerate(em):
            if k==0:
                if j>threshold and j<upper and h1[k]>1000:
                    base_h[i]=k
                    extin_h[i]=j
                    ht[i]=h1[k]
                    break
            elif k==len(em)-1:
                if j>threshold and j<upper and j>em[k-1] and h1[k]>1000:
                    base_h[i]=k
                    extin_h[i]=j
                    ht[i]=h1[k]
                    break
            else:
                if j>threshold and j<upper and j>em[k-1] and j<em[k+1] and h1[k]>1000:
                    base_h[i]=k
                    extin_h[i]=j
                    ht[i]=h1[k]
                    lower_extin[i]=em[k-1]
                    upper_extin[i]=em[k+1]
                    break
    return base_h,lower_extin,extin_h,upper_extin,ht
```

```python
def ccn_cal(ccn,sup):
    x=np.where((sup>=0.95) & (sup<=1.05))[0]    
    uu=np.zeros(sup.shape)
    for i in range(len(ccn)):
        uu[i]=(np.argmin(np.abs(x-i)))
    return [ccn[int(p)] for p in uu]
```

```python
def data_gen(date):
    start=time.time()
    print(f'<--------Generating data for {date}---------->')
    k1,m1,r1,s1,e1,mp1=generate_cdfs(date)
    tk=np.ma.filled(k1['time'][:])
    hk=np.ma.filled(k1['height'][:])
    tm=np.ma.filled(m1['time'][:])
    hm=np.ma.filled(m1['height'][:])
    tr=np.ma.filled(r1['time'][:])
    hr=np.ma.filled(r1['height'][:])*1000
    ts=np.ma.filled(s1['time'][:])
    te=np.ma.filled(e1['time_offset'][:])
    he=np.ma.filled(e1['height_high'][:])*1000 
    tmp=np.ma.filled(mp1['time'][:])
    hmp=np.ma.filled(mp1['height'][:])*1000
    ext=np.ma.filled(mp1['backscatter'][:])*18.8
    exr=r1['ext'][:]
    sup=s1['supersaturation_calculated'][:]
    ccn=np.ma.filled(s1['N_CCN'][:])
    temp=np.ma.filled(r1['temperature'][:])+273.15
    print('Generating indices for extinction')
    e_ind,e_low,ex,e_high,h=extinction(ext,tmp,hmp)
    targs,hargs=st_clustering(tmp,hmp,tk,hk)
    lwc,dlwc=data_gen_par(targs,hargs,m1['liquid_water_content'],e_ind)
    v,dv=data_gen_par(targs,hargs,k1['mean_doppler_velocity'],e_ind)
    sw,dsw=data_gen_par(targs,hargs,k1['spectral_width'],e_ind)
    ref,dref=data_gen_par(targs,hargs,k1['reflectivity'],e_ind)
    dat=[date]*len(tmp)
    print('Processing temperature')
    targs2,hargs2=st_clustering(tmp,hmp,tr,hr)
    e1,e2,e3,e4,e5=extinction(exr,tr,hr)
    tem=[]
    for tt in targs2:
        tem.append(temp[tt,int(e1[tt])])
    print('Processing CCN')
    targs3,hargs3=st_clustering(tmp,hm,ts,he)
    ccn2=[]
    sup2=[]
    for x in targs3:
        ccn2.append(ccn[int(x)])
        sup2.append(sup[int(x)])
    ccn_e=ccn_cal(np.array(ccn2),np.array(sup2))
    print('Generating Output file')
    lll=[dat,tmp,h,lwc,dlwc,v,dv,sw,dsw,ref,dref,tem,e_low,ex,e_high,ccn_e]
    var=['Date','Time','Height','LWC','LWC_SD','Velocity','Velocity_SD','Spectral_Width','Spectral_Width_SD','Reflectivity','Reflectivity_SD','Temperature','Extinction_low','Extinction','Extinction_high','CCN']
    df={}
    for uu in range(len(lll)):
        print(var[uu],len(lll[uu]))
        df.update({var[uu]:lll[uu]})
    tim1=time.time()
    llll=tim1-start
    print(f'Time taken so far {llll:1.2f}')
    return pd.DataFrame(df)
```
